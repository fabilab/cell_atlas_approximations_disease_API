#### **Import and initialise api**

```python
import atlasapprox_disease
api = atlasapprox_disease.API()
```

```python
cell_metadata_lung = api.metadata(
    tissue = "lung",
    development_stage = "adult"
)

cell_metadata_lung
```

### **Querying differential cell type abundance**
The ``differential_cell_type_abundance`` function allows you to get differential cell type abundance between two conditions across datasets. (e.g., disease vs normal, male vs female). The following example show yoou how to browser a overview of differential abundance of all cell types in diabetes



```python
df = api.differential_cell_type_abundance (
    disease = "diabetes",
    tissue = "pancreas"
)
df
```

```python
# Create the up-down bar chart (rotated)
# plt.figure(figsize=(7, 3))
sns.barplot(
    x='cell_type',
    y='delta_frac',
    data=df,
    hue=df['delta_frac'] > 0,  # Color bars based on positive/negative delta_frac
    palette={True:'#69c0ff', False:'#8c8c8c'}
)

# Customize the chart
plt.ylabel('Change in fraction (Diabetes - Normal)')
plt.xlabel('Cell Type')
plt.title('Differential cell type Abundance in diabetes (pancreas)')
# plt.axhline(0, color='black', linewidth=0.8)  # Add a horizontal line at y=0

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Update the legend
plt.legend(
    title='Change',
    labels=['Decrease', 'Increase'],
    handles=[
        plt.Rectangle((0,0),1,1, color='#69c0ff'),
        plt.Rectangle((0,0),1,1, color='#8c8c8c')
    ]
)

plt.tight_layout()
```

### **Querying average gene expression**
The ``average`` function allows you to retrieve average expression of your gene of interest in specified disease or cell type. Give you a rough idea of whether a gene is expressed in what cell_type, tissue and disease.


```python
api.average (
    features = "ACE2,TLR4,NLRP3,MBL2,IL6",
    disease = "Covid"
)
```

```python
dotplot_data = api.dotplot(
    features = "ACE2,TMPRSS2",
    disease = "Covid"
)

dotplot_data
```
